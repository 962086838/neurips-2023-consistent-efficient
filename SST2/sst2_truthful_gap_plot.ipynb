```python
print("Harmonica 2 order")
degree = 2
idx = 0

for C in [1,2,3]:
    print("C =", C)
    for each_subspace_num in subspace_num:
        split_start = 0; split_end = 218
        final_answer_output = np.load(f"./harmonica2degree/harmonica2degree_truthful_answer_C_{C}_subspace_{each_subspace_num}_seed123_{split_start}_{split_end}.npy", allow_pickle=True)
        for split_start, split_end in zip([218, 436, 654], 
                                          [436, 654, 872]):
            temp_answer_output = np.load(f"./harmonica2degree/harmonica2degree_truthful_answer_C_{C}_subspace_{each_subspace_num}_seed123_{split_start}_{split_end}.npy", allow_pickle=True)
            final_answer_output = np.concatenate([final_answer_output, temp_answer_output], axis = 0)
        
        print(final_answer_output.shape)
        untruthfulness = 0

        for _answer in final_answer_output:
            untruthfulness += np.sum(_answer ** 2)


        if (each_subspace_num == 0):
            print(f"subspace infinity")
        else:
            print(f"subspace {each_subspace_num}")
        
        print("Untruthfulness", np.round(untruthfulness / len(final_answer_output), 4))
        
        eval(f"C{C}_untruthfulness_harmonica")[0][subspace_num.index(each_subspace_num)] = untruthfulness / len(final_answer_output)
        eval(f"C{C}_untruthfulness")[0][subspace_num.index(each_subspace_num)] = untruthfulness / len(final_answer_output)
        idx += 1
```

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt

index = ["1","2","4","8","16","32","∞"] 

for C in [1,2,3]:
    eval(f"C{C}_untruthfulness")
    plt.figure()
    df_L2 = pd.DataFrame(eval(f"C{C}_untruthfulness").T, index, ["Harmonica (2 order)", "Harmonica (3 order)", "LIME", "SHAP", "Integrated Gradients", "Integrated Hessians", "Shapley Taylor Index (2 order)"]) 

    ax_L2 = sns.lineplot(data=df_L2, marker='X') 
    # ax_L2.set(xlabel='Subspace radius', ylabel='Readable inconsistency') 
    # ax_L2.set(yscale='log')
    # ax_L2.set(ylim=[0,1])
    fig_L2 = ax_L2.get_figure() 
    ax_L2.legend(fontsize=11)
    plt.xlabel('Subspace radius', fontsize=17)
    plt.ylabel('Truthful gap', fontsize=17)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.tight_layout()
    fig_L2.savefig(f"truthful-gap-imdb-plot-c{C}.pdf")
```
